```python
import numpy as np
import pandas as pd

world = pd.read_csv('world.csv')
world.info() #上面資料的說明書
```

```python
world1 = world.reindex(columns=['洲名','國家','日期','總確診數','新增確診數','總人口數','解封指數'])
world1
```

```python
taiwan = world1[world1["國家"] == '台灣']
taiwan
```

```python
japan = world1[world1["國家"] =='日本']
japan
```

```python
taiwan.query('日期>="2021-01-01" and 日期<="2021-01-31"')
```
