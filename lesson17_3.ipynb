```python
import numpy as np
import pandas as pd
import requests

#每日進站人口數,政府開放平台 - https://data.gov.tw/dataset/8792
url1 = 'https://ods.railway.gov.tw/tra-ods-web/ods/download/dataResource/8ae4cabf6973990e0169947ed32454b9'


#車站基本資料集 - https://data.gov.tw/dataset/33425
url2 = 'https://ods.railway.gov.tw/tra-ods-web/ods/download/dataResource/0518b833e8964d53bfea3f7691aea0ee'

response = requests.get(url1)
sites = response.json()
sites
```

```python
#車站基本資料集 - https://data.gov.tw/dataset/33425
url2 = 'https://ods.railway.gov.tw/tra-ods-web/ods/download/dataResource/0518b833e8964d53bfea3f7691aea0ee'

response = requests.get(url2)
site_info = response.json()
site_info
import pandas as pd
site_entry_df = pd.DataFrame(sites)
site_entry_df.columns = ['乘車日','車站代碼','進站人數','出站人數']
site_entry_df
```

```python
site_info_df = pd.DataFrame(site_info,columns=["stationCode","stationName"])
site_info_df.columns = ["車站代碼","中文站名"]
site_info_df
```

```python
pd.merge(site_entry_df,site_info_df,how="left",left_on="車站代碼",right_on="車站代碼")
```

```python
sites_df = pd.merge(site_entry_df,site_info_df,how="left",left_on="車站代碼",right_on="車站代碼")
sites_df1 = sites_df.reindex(columns=["乘車日","中文站名","進站人數","出站人數"])
sites_df1
```

```python
sites_df2 = sites_df1.set_index('中文站名')
sites_df2
```

```python
sites_df2.info()
```

```python
def toInt(value:str)->int:
    return int(value)
sites_df2['進站人數']= sites_df2['進站人數'].map(toInt)
```

```python
sites_df2.info()
```

```python
def toInt(value:str)->int:
    try:
        return int(value)
    except:
        return 0
#sites_df2['進站人數']= sites_df2['進站人數'].map(lambda value:int(value))
sites_df2['進站人數']= sites_df2['進站人數'].map(toInt)
```

```python
sites_df2['出站人數'] = sites_df2['出站人數'].astype(int)
```

```python
sites_df2.info()
```

```python
sites_df3 = sites_df2.groupby('中文站名')[['進站人數','出站人數']].mean()
sites_df3
```
