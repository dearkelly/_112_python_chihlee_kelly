```python
import numpy as np
import pandas as pd
import requests

#, - https://data.gov.tw/dataset/8792
url1 = 'https://ods.railway.gov.tw/tra-ods-web/ods/download/dataResource/8ae4cabf6973990e0169947ed32454b9'


# - https://data.gov.tw/dataset/33425
url2 = 'https://ods.railway.gov.tw/tra-ods-web/ods/download/dataResource/0518b833e8964d53bfea3f7691aea0ee'

response = requests.get(url1)
sites = response.json()
sites
```

```python
# - https://data.gov.tw/dataset/33425
url2 = 'https://ods.railway.gov.tw/tra-ods-web/ods/download/dataResource/0518b833e8964d53bfea3f7691aea0ee'

response = requests.get(url2)
site_info = response.json()
site_info
import pandas as pd
site_entry_df = pd.DataFrame(sites)
site_entry_df.columns = ['','','','']
site_entry_df
```

```python
site_info_df = pd.DataFrame(site_info,columns=["stationCode","stationName"])
site_info_df.columns = ["",""]
site_info_df
```

```python
pd.merge(site_entry_df,site_info_df,how="left",left_on="",right_on="")
```

```python
sites_df = pd.merge(site_entry_df,site_info_df,how="left",left_on="",right_on="")
sites_df1 = sites_df.reindex(columns=["","","",""])
sites_df1
```

```python
sites_df2 = sites_df1.set_index('')
sites_df2
```

```python
sites_df2.info()
```
