```python
import requests
from bs4 import BeautifulSoup

url = "https://www.taiwanlife.com/product/7/T07N1"
response = requests.get(url)
response.encoding = 'utf-8'
if response.ok:
    print("下載成功")
    print(response.text)
else:
    print("下載失敗")
```

```python
bs4 = BeautifulSoup(response.text,'html.parser')
result1 = bs4.find_all(name ="ul",class_= 'editor-element')
for ul in result1:
    result2 = ul.find_all(name='li')
    for liTag in result2:
        print(liTag.string)
```
