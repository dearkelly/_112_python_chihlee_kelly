```python
import numpy as np
import pandas as pd
scroes = np.random.randint(50, 100,size=(50,5))
scores_df = pd.DataFrame(scroes,
             columns=['國文', '英文', '數學', '地理', '歷史'],
             index=range(1,51))
scores_df
sum_values = scores_df.sum(axis=1)
mean_values = scores_df.mean(axis=1)
scores_df['總分'] = sum_values
scores_df['平均'] = mean_values
```

```python
scores_df.sort_values(by='總分')#排序總分
```

```python
scores_df.sort_values(by='總分',ascending=True).iloc[:5] #排序後5名總分
```

```python
scores_df.sort_values(by='總分',ascending=False).iloc[:5] #排序前5名總分
```

```python
#使用SUBSCRIPT()過瀘資料
scores_df[scores_df['平均'] < 80]
```

```python
#使用SUBSCRIPT()過瀘資料
scores_df[scores_df['平均'] < 80].count()
```

```python
#平均80以下，歷史不及格 使用subscrip[]的搜尋方法
scores_df[(scores_df['平均'] < 80) & (scores_df['歷史'] < 60)]
```

```python
#使用query(字串)方法
scores_df.query('平均<80')
```

```python
scores_df.query('平均<80 and 歷史<60')
```
