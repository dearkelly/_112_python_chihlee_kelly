```python
import numpy as np #2維資料
import pandas as pd

df = pd.DataFrame(np.arange(12).reshape((4,3)),
                index=[['a','a','b','b'],[1,2,1,2]],
                columns = [['台北','台北','台中'],['Green','Red','Green']])

df
```

```python
df.index.names = ["key1","key2"]
df.columns.names = ["縣市","顏色"]
df
```

```python
df['台北']['Green']
```

```python
df.loc['a']['台中']
```

```python
df.swaplevel('key1','key2',axis=0) #key1,key2對調
```

```python
df.swaplevel('縣市','顏色',axis=1)
```

```python
df.sort_index(level='key2')
```
